```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split



df = pd.read_csv('canada_per_capita_income.csv')
df.head()


X=df[['year']]
y=df['per capita income (US$)']


X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.4)
X_train.tail()


y_train.head()

X_test.head()

y_test.head()

regressor=LinearRegression()
regressor.fit(X_train,y_train)
y_pred = regressor.predict(X_test)



plt.scatter(X_train,y_train,color='g')
plt.plot(X_train , regressor.predict(X_train),color='r')
plt.title('Income vs Year (Training set)')
plt.xlabel('Year')
plt.ylabel('Income')
plt.show()



plt.scatter(X_test, y_test, color = 'g')
plt.plot(X_train, regressor.predict(X_train), color = 'r')
plt.title('Income vs Year (Test set)')
plt.xlabel('Year')
plt.ylabel('Income')
plt.show()



plt.scatter(X_test, y_test, color = 'g')
plt.plot(X_test, y_pred, color = 'r')
plt.title('Income vs Year (Test set)')
plt.xlabel('Year')
plt.ylabel('Income')
plt.show()



prediction_for_2020 = regressor.predict([[2020]])
print (prediction_for_2020)
print(f"Accuracy:  {regressor.score(X_test,y_test)*100 :.2f}%")


coefficient = regressor.coef_
intercept = regressor.intercept_
print ("Coefficient:",coefficient)
print("Slope/Intercept:" , intercept)
```

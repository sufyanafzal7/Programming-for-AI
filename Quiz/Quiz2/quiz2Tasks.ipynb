Q.2 Data Summary and Statistics

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
# print(df.head())
grouped = df.groupby('target')
samplesPerSpecies = grouped.size()
print("Number of samples per species:\n", samplesPerSpecies)
meanPerSpecies = grouped.mean()
stdPerSpecies = grouped.std()
print("\nMean of features grouped by species:\n", meanPerSpecies)
print("\nStandard deviation of features grouped by species:\n", stdPerSpecies)
```

Q.3 Correlation Analysis

```python
import pandas as pd
from sklearn.datasets import load_iris
import seaborn as sns
import matplotlib.pyplot as plt
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
# print(df.head())
grouped = df.groupby('target')
samplesPerSpecies = grouped.size()
# print("Number of samples per species:\n", samplesPerSpecies)
meanPerSpecies = grouped.mean()
stdPerSpecies = grouped.std()
# print("\nMean of features grouped by species:\n", meanPerSpecies)
# print("\nStandard deviation of features grouped by species:\n", stdPerSpecies)
correlationMatrix = df.iloc[:, :-1].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlationMatrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Numeric Features")
plt.show()
print("\nCorrelation Matrix:\n", correlationMatrix)
```

Q.4 Outlier Detection

```python
import pandas as pd
from sklearn.datasets import load_iris
import seaborn as sns
import matplotlib.pyplot as plt
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
# print(df.head())
grouped = df.groupby('target')
samplesPerSpecies = grouped.size()
# print("Number of samples per species:\n", samplesPerSpecies)
meanPerSpecies = grouped.mean()
stdPerSpecies = grouped.std()
# print("\nMean of features grouped by species:\n", meanPerSpecies)
# print("\nStandard deviation of features grouped by species:\n", stdPerSpecies)
numericalFeatures = df.columns[:-1]
plt.figure(figsize=(12, 8))
for i, feature in enumerate(numericalFeatures, 1):
    plt.subplot(2, 2, i)  
    sns.boxplot(data=df, x='target', y=feature)
    plt.title(f"Boxplot of {feature}")
    plt.xlabel("Species")
    plt.ylabel(feature)

plt.tight_layout()
plt.show()
```

Q.5 Pairwise Relationship

```python
import pandas as pd
from sklearn.datasets import load_iris
import seaborn as sns
import matplotlib.pyplot as plt
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
# print(df.head())
grouped = df.groupby('target')
samplesPerSpecies = grouped.size()
# print("Number of samples per species:\n", samplesPerSpecies)
meanPerSpecies = grouped.mean()
stdPerSpecies = grouped.std()
# print("\nMean of features grouped by species:\n", meanPerSpecies)
# print("\nStandard deviation of features grouped by species:\n", stdPerSpecies)
sns.pairplot(df, hue="target", diag_kind="hist", palette="viridis")
plt.suptitle("Pairwise Relationships of Iris Dataset", y=1.02)
plt.show()
```

Question 2 : Visulization

-----------

a. Plot the distribution of passengers' ages using a histogram.

```python
import pandas as pd
import matplotlib.pyplot as plt
titanicData = pd.read_csv('Titanic-Dataset.csv')
plt.hist(titanicData['Age'].dropna(), bins=20, edgecolor='black')
plt.title("Distribution of Passengers' Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

b. Create a countplot of passengers by class ('class').

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
titanicData = pd.read_csv('Titanic-Dataset.csv')
sns.countplot(data=titanicData, x='Pclass')
plt.title("Count of Passengers by Class")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

c. Create a box plot showing the fare distribution across different classes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
titanicData = pd.read_csv('Titanic-Dataset.csv')
sns.boxplot(data=titanicData, x='Pclass', y='Fare')
plt.title("Fare Distribution Across Passenger Classes")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

d. Plot a heatmap of the correlation matrix.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
titanicData = pd.read_csv('Titanic-Dataset.csv')
correlationMatrix = titanicData.corr(numeric_only=True)
plt.figure(figsize=(10, 6))
sns.heatmap(correlationMatrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

Question 3 : Grouping and Aggregation

---------------

a. Find the avarage fare paid by each class of passengers.

```python
import pandas as pd
titanicData = pd.read_csv('Titanic-Dataset.csv')
averageFareByClass = titanicData.groupby('Pclass')['Fare'].mean()
print(averageFareByClass)
```

b. Calculate the survival rate ('mean of survived') grouped by gender ('sex').

```python
import pandas as pd
titanicData = pd.read_csv('Titanic-Dataset.csv')
survivalRateByGender = titanicData.groupby('Sex')['Survived'].mean()
print(survivalRateByGender)
```

c. Show the number of passengers grouped by class and gender.

```python
import pandas as pd
titanicData = pd.read_csv('Titanic-Dataset.csv')
passengerCount = titanicData.groupby(['Pclass', 'Sex']).size()
print(passengerCount)
```

Question 4 : Custom Task - EDA on 'tips' dataset

--------------

a. Find the avarage tip given by smokers vs non-smokers.

```python
import pandas as pd
tipsData = pd.read_csv('tips.csv')
averageTips = tipsData.groupby('smoker')['tip'].mean()
print(averageTips)
```

b. Plot a scatter plot showing 'total_bill' vs 'tip', colored by 'sex'.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
tipsData = pd.read_csv('tips.csv')
sns.scatterplot(data=tipsData, x='total_bill', y='tip', hue='sex')
plt.title("Total Bill vs Tip by Sex")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```
